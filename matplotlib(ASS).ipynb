```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:/Users/gunap/Downloads/Virat Kohli DataSet.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Start Date']=pd.to_datetime(df['Start Date'])
#to extract only year from the entire start date
df['year']=df['Start Date'].dt.year
```

```python
df.info()
```

```python
#to reduce too many bars use groupby
year_runs=df.groupby('year')['Runs'].sum()
```

```python
df.sort_values('Start Date').head()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,6),dpi=100)
#Runs scored over years
axes[0].bar(df['year'],df['Runs'])
axes[0].set_xlabel('years')
axes[0].set_ylabel('Runs');
axes[0].set_title('Runs over Years')
plt.tight_layout()#do not overlap

#Batting avg over years
axes[1].bar(df['year'],df['BF'])
axes[1].set_xlabel('years')
axes[1].set_ylabel('BF');
axes[1].set_title('Batting avg Years')
plt.tight_layout()#do not overlap
plt.show()#end of the picture
```
