<h4> <b> Statistical Operations </b> </h4>

```python
# Creating a New Array
new_array=np.array([[ 2,  3,  4],
       [ 4,  3, 17],
       [89, 33, 12]])
new_array
```

```python
# Sum Operations
# Produce total sum, row wise sum (first elements), columns wise sum

col_sum=np.sum(new_array,axis=0)
col_sum

row_sum=np.sum(new_array,axis=1)
row_sum

total_sum=np.sum(new_array)
total_sum

result=(total_sum,col_sum,row_sum)
print(result)
```

```python
# Calculate Mean, Meadian, Square root, minimum, maximum values of new_arr
arr_mean=np.mean(new_array)
arr_mean

arr_median=np.median(new_array)
arr_median

arr_sqrt=np.sqrt(new_array)
arr_sqrt

arr_min=np.min(new_array)
arr_min

arr_max=np.max(new_array)
arr_max

result=(arr_mean,arr_median,arr_sqrt,arr_min,arr_max)
print(result)
```
