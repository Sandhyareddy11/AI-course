```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Plot a scatterplot to analyze the relationship between Unit Price and Customer Rating based on Gender
#Plot the distribution of Sales_Amount with a histogram and KDE curve
#Show the number of purchases made in each Region with a Count Plot
#Plot the Distribution of Sales_Amount for each Product
```

```python
data=pd.read_csv(r"C:/Users/gunap/OneDrive/Desktop/data sets/sales_data.csv")
data.head()
```

```python
#Plot a scatterplot to analyze the relationship between Unit Price and Customer Rating based on Gender
sns.scatterplot(data=data,x="Unit_Price",y="Customer_Rating",hue="Gender")
plt.title("Unit Price  vs  Customer Rating by Gender")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Plot the distribution of Sales_Amount with a histogram and KDE curve
plt.figure(figsize=(8,5))
sns.histplot(data=data, x="Sales_Amount", kde=True)
plt.title("Distribution of Sales Amount")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
#Show the number of purchases made in each Region with a Count Plot
plt.figure(figsize=(6,5))
sns.countplot(data=data, x="Region")
plt.title("Number of Purchases in Each Region")
plt.xlabel("Region")
plt.ylabel("Count of Purchases")
plt.tight_layout()
plt.show()
```

```python
#Plot the Distribution of Sales_Amount for each Product
plt.figure(figsize=(8,5))
sns.displot(data=data, x="Sales_Amount", hue="Product", kde=True)
plt.title("Distribution of Sales Amount for Each Product")
plt.tight_layout()
plt.show()
```
