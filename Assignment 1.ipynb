<h6>1)A company wants to calculate the final salary of an employee after adding a bonus. The 
bonus depends on the employee’s years of experience and monthly salary. 

Bonus Rules: 

• If the employee has more than 5 years of experience and the salary is less than 
50,000/- : Add 10% bonus to the salary 
• If the employee has 3 to 5 years of experience: Add 5% bonus to the salary 
• If the employee has less than 3 years of experience: No bonus 

Tasks to Perform: 
1. Take the following inputs from the user: 
Employee Name, Monthly Salary, Years of Experience 
2. Calculate the bonus amount 
3. Calculate the final salary 
4. Display: 
Employee Name, Experience, Bonus Amount, Final Salary 

<h6>Requirements: 

• Use a function to calculate the bonus 
• Use conditional statements to apply bonus rules 
• Use proper variables and data types 
• Print output
</h6>

```python
name=input("Employee Name:")
sal=int(input("Monthly Salary:"))
exp= int(input("Years of Experience:"))
def bonus(sal,exp):
  if exp>5 and sal<50000:
     return 10/100*sal
  elif exp>=3 and exp<=5:
     return 5/100*sal
  else:
     return 0
bonus_amount=bonus(sal,exp)
final_sal=sal + bonus_amount
print("Final Salary:",final_sal)
```

<h6>2)A training institute wants to decide whether a student is eligible to receive a certificate. 
The eligibility depends on the student’s attendance percentage and marks obtained in 
subjects. 
    
Eligibility Rules: 
    
A student is Eligible for Certification only if both conditions are satisfied: 
• Attendance must be at least 75% 
• Marks in all subjects must be 40 or above 
• If any one of the above conditions is not satisfied, the student is Not Eligible. 
    
Tasks to Perform:
    
1. Take the following details as input from the user: 
Student Name, Total number of classes conducted Number of classes 
attended 
2. Calculate the attendance percentage 
3. Take marks of 3 subjects using a loop 
4. Check if the student meets both eligibility conditions 
5. Display: 
Student Name, Attendance Percentage, Marks of all subjects, Final Eligibility Status 
(Eligible / Not Eligible) 
    
Program Requirements: 
    
• Use a function to check eligibility 
• Use loops to accept subject marks 
• Use conditional statements for validation 
• Use proper variables and data types 
• Print output</h6>

```python
name = input("Student Name: ")
conducted = int(input("Total number of classes conducted: "))
attended = int(input("Total number of classes attended: "))

attendance_percentage = (attended / conducted) * 100

def check_eligibility(attendance_percentage, student_marks):
    if attendance_percentage >= 75 and min(student_marks) >= 40:
        return "Eligible"
    else:
        return "Not Eligible"

student_marks = []
for i in range(1, 4):
    mark = int(input("Enter marks of subject:" ))
    student_marks.append(mark)

result = check_eligibility(attendance_percentage, student_marks)
print("Eligibility Status:", result)
```
