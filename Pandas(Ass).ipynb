<h3>Part 3: Data Cleaning </h3>


```python
#Task 1: Remove the S N column from the DataFrame. 
#Task 2: Fill missing values in Inches column using mode. 
# Task 3: Fill missing values in Ram column using mode. 
#Task 4: Verify missing values handling using df.info() 
#Task 5: Convert Inches column to numeric type. 
#Task 6: Convert Ram column to numeric by removing text (GB). 
#Task 7: Convert Weight column to numeric by removing text (kg) 
```

```python
df=df.drop('S N',axis=1)
df.head()
```

```python
df['Inches']=df['Inches'].fillna(df['Inches'].mode()[0])
df['Inches']
```

```python
df['Ram']=df['Ram'].fillna(df['Inches'].mode()[0])
df['Ram']
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Convert Inches column to numeric type. 
df["Inches"]=pd.to_numeric(df["Inches"].replace(r"?",regex=True))
```

```python
#Convert Ram column to numeric by removing text (GB).
df["Ram"]=pd.to_numeric(df["Ram"].astype(str).str.replace(r"[^0-9.]","",regex=True))
```

```python
# Task 7: Convert Weight column to numeric by removing text (kg) 
df["Weight"]=pd.to_numeric(df["Weight"].astype(str).str.replace(r"[^0-9.]","",regex=True))
```

<h3>Part5:Creating New Columns</h3>

```python
#Task 1: Create a new column ‘Price in $’ by converting Price column into $ Dollars. 
#Task 2: Create Weight_Category column based on laptop weight. 
#“Light” → Weight < 2kg 
#“Heavy” → Weight ≥ 2kg
```

```python
#Task 1: Create a new column ‘Price in $’ by converting Price column into $ Dollars. 
usd_rate=83
df["Price in $"]=(df["Price"]/usd_rate)
```

```python
#Task 2: Create Weight_Category column based on laptop weight. 
#“Light” → Weight < 2kg 
#“Heavy” → Weight ≥ 2kg
import numpy as np
df["Weight_Category"]=np.where(df["Weight"]< 2,  "Light","Heavy")
```
